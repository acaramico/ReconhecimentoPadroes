```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
from sklearn import metrics

from sklearn.datasets import load_iris
X, y = load_iris(return_X_y = True, as_frame = True)

X = X.loc[:,['petal length (cm)', 'petal width (cm)']]
```

```python
fig, ax = plt.subplots()
ax.scatter(X['petal length (cm)'], X['petal width (cm)'])
plt.show()
```

```python
kmeans = KMeans(n_clusters = 2, n_init = 'auto').fit(X)
kmeans.labels_
```

```python
fig, ax = plt.subplots()
ax.scatter(X['petal length (cm)'], X['petal width (cm)'], c = kmeans.labels_)
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, n_init = 'auto').fit(X)
kmeans.labels_
```

```python
fig, ax = plt.subplots()
ax.scatter(X['petal length (cm)'], X['petal width (cm)'], c = kmeans.labels_)
plt.show()
```

```python
valores_k = []
inercia =[]

for i in range(1, 15):
    kmeans = KMeans(n_clusters = i, n_init = 'auto').fit(X)
    valores_k.append(i)
    inercia.append(kmeans.inertia_)
    print(kmeans.inertia_)
```

```python
fig, ax = plt.subplots()
ax.plot(valores_k, inercia)
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, n_init='auto')
kmeans.fit(X)  # Fit the model to your data

# Access the centroids
centroids = kmeans.cluster_centers_

# Create the plot
fig, ax = plt.subplots(ncols=2, figsize=(15, 6))

# Scatter plot for KMeans clustering
ax[0].scatter(X['petal length (cm)'], X['petal width (cm)'], c=kmeans.labels_)
ax[0].scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, color='red')  

# Scatter plot for the true labels
ax[1].scatter(X['petal length (cm)'], X['petal width (cm)'], c=y)

# Show the plot
plt.show()
```

```python
valores_k = []
ARI = []
RI = []

for i in range(1, 15):
    kmeans = KMeans(n_clusters = i, n_init = 'auto').fit(X)
    valores_k.append(i)
    RI.append(metrics.rand_score(y, kmeans.labels_))
    ARI.append(metrics.adjusted_rand_score(y, kmeans.labels_))

fig, ax = plt.subplots()
ax.plot(valores_k, RI)
ax.plot(valores_k, ARI)
plt.show()
```

```python
valores_k = []
S = []

for i in range(2,15):
    kmeans = KMeans(n_clusters = i, n_init = 'auto').fit(X)
    valores_k.append(i)
    S.append(metrics.silhouette_score(X, kmeans.labels_))
```

```python
fig, ax = plt.subplots()
ax.plot(valores_k, S)
plt.show()
```

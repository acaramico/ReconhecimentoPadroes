```python
import numpy as np
import cv2
import matplotlib as plt 
%matplotlib inline 
import matplotlib.pyplot as plt 

imagem = cv2.imread('pessoa_sorrindo.jpg')
imagemGray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

smile = cv2.CascadeClassifier('haarcascade_smile.xml')
face_frontal = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

smile_recon = smile.detectMultiScale(imagemGray, scaleFactor = 2.9, minNeighbors = 5)
face_recon = face_frontal.detectMultiScale(imagemGray, scaleFactor = 2.0, minNeighbors = 1)

print('sorrisos encontrados - ', len(smile_recon))
print('faces encontradas - ', len(face_recon))

for(x, y, w, h) in face_recon:
    cv2.rectangle(imagem, (x,y), (x+w, y+h), (0, 255, 0), 4)

for(x, y, w, h) in smile_recon:
    cv2.rectangle(imagem, (x,y), (x+w, y+h), (255, 0, 0), 4)

plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
```

```python
imagem = cv2.imread('pessoas_sorrindo.jpg')
imagemGray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

smile = cv2.CascadeClassifier('haarcascade_smile.xml')
face_frontal = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

smile_recon = smile.detectMultiScale(imagemGray, scaleFactor = 9, minNeighbors = 5)
face_recon = face_frontal.detectMultiScale(imagemGray, scaleFactor = 2.0, minNeighbors = 1)

print('sorrisos encontrados - ', len(smile_recon))
print('faces encontradas - ', len(face_recon))

for(x, y, w, h) in face_recon:
    cv2.rectangle(imagem, (x,y), (x+w, y+h), (0, 255, 0), 4)

for(x, y, w, h) in smile_recon:
    cv2.rectangle(imagem, (x,y), (x+w, y+h), (255, 255, 0), 4)

plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
```

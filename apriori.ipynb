```python
import numpy as np 
import matplotlib as plt 
import pandas as pd 
from apyori import apriori
```

```python
df = pd.read_csv("store_data.csv", delimiter= ',', header= None)
```

```python
df.replace(np.nan, 0, inplace=True)
df.head(10)
```

```python
def removeZeros(lista):
    return list(filter(lambda x: x != 0, lista))
```

```python
listao = []
for index, product in df.iterrows():
    lista = product.values.tolist()
    lista = removeZeros(lista)
    listao.append(lista)
```

```python
print(lista)
```

```python
regras = apriori(transactions= listao, min_support= 0.002, min_confidence= 0.02, min_lift= 2)
```

```python
final = list(regras)
```

```python
def fim(final):
    X = [tuple(result[2][0][0])[0] for result in final]
    Y = [tuple(result[2][0][1])[0] for result in final]
    sup = [result[1] for result in final]
    conf = [result[2][0][2] for result in final]
    lif = [result[2][0][3] for result in final]
    return list (zip(X, Y, sup, conf, lif))
    
```

```python
resultDF = pd.DataFrame(fim(final), columns= ['X', 'Y', 'Suporte', 'Confianca', 'Lift'])
```

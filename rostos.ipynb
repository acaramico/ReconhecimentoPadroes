```python
import numpy as np
import cv2
import matplotlib as plt
%matplotlib inline
import matplotlib.pyplot as plt 

imagem = cv2.imread('selecao.jpg')
imagemGray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

face_frontal = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

face_recon = face_frontal.detectMultiScale(imagemGray, scaleFactor = 1.3, minNeighbors = 1)

for(x, y, w, h) in face_recon:
    cv2.rectangle(imagem, (x,y), (x+w, y+h), (0, 255, 0), 4)

plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
```

```python
import numpy as np
import cv2
import matplotlib as plt
%matplotlib inline
import matplotlib.pyplot as plt 

imagem = cv2.imread('Gatos2.jpg')
imagemGray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

face_frontal = cv2.CascadeClassifier('haarcascade_frontalcatface.xml')

face_recon = face_frontal.detectMultiScale(imagemGray, scaleFactor = 1.1, minNeighbors = 1, maxSize= (60,60))

for(x, y, w, h) in face_recon:
    cv2.rectangle(imagem, (x,y), (x+w, y+h), (0, 255, 0), 4)

plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
```

```python
import cv2
import mediapipe as mp

cam = cv2.VideoCapture(0)
Reconhecimento = mp.solutions.face_detection
Reconhecedor = Reconhecimento.FaceDetection()
Draw = mp.solutions.drawing_utils

while True:
    status, image = cam.read()
    if not status:
        break
    faces = Reconhecedor.process(image)
    
if faces.detections:
    for rosto in faces.detections:
        Draw.draw_detection(image, rosto)

cv2.imshow("Faces na webcam", image)

if cv2.waitKey(5) == 27:
    break 

cam.release()
cv2.destroyAllWindows()
```
